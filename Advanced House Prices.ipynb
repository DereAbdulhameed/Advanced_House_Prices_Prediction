###  Separate the dataframe into Object and Numerical Datatypes for easy exploration

```python
object_datatypes = df.select_dtypes(include = 'object')
object_datatypes.head(2)
```

```python
# Fill in missing values in all columns of the object_datatype
```

```python
for i in object_datatypes.columns:
    a = object_datatypes[i].mode()[0]
    object_datatypes[i].fillna(a, inplace = True)
```

```python
# Inspect
object_datatypes.isnull().sum()
```

### Drop outliers in Object Dataframe

### Convert Object datatype into numerical using SKLearn

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
```

```python
object_datatypes['Street'] = lb.fit_transform(object_datatypes['Street'])
```

```python
object_datatypes['Street'].dtype
```

```python
for i in object_datatypes.columns:
    object_datatypes[i] = lb.fit_transform(object_datatypes[i])
    
```

```python
# Assert is used to verify is a value is equal to another value
assert len(object_datatypes) == len(numerical_datatypes)
```

### Concatenate both dataframes 
This is after missing values have been handled as well as encoding of categorical data

```python
frames = [object_datatypes, numerical_datatypes]
```

```python
df2 = pd.concat(frames, axis = 1)
```

```python
df2.head(2)
```

### Splitting the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
feature = df2.iloc[:,:-1]
# Another method is feature = df2.drop(['SalePrice'])
target = df2.iloc[:,-1]
# Another method is target = df2['SalePrice']
```

```python
feature.head(2)
```

```python
target.head(2)
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(feature, target,
                                                                          test_size=0.2, random_state=2021)


print(feature_train.shape)
print(feature_test.shape)
print(target_train.shape)
print(target_test.shape)
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
scale = StandardScaler()
feature_train_scaled = scale.fit_transform(feature_train)
feature_test_scaled = scale.transform(feature_test)
```

```python
feature_test_scaled[:1]
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg = reg.fit(feature_train_scaled, target_train)
```

```python
target_predicition = reg.predict(feature_test_scaled)
```

```python
target_predicition[:5]
```

```python
target_test[:5]
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
mean_squared_error(target_test, target_predicition, squared = False)
```

```python
# GradientBoostClasifier
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(random_state=0)
```

```python
gbr.fit(feature_train_scaled, target_train)
```

```python
target_predicition = gbr.predict(feature_test_scaled)
```

```python
target_predicition[:5]
```

```python
target_test[:5]
```

```python
mean_squared_error(target_test, target_predicition, squared = False)
```
